```python
import pandas as pd

# Load cleaned dataset

df = pd.read_csv("../Outputs/cleaned_dataset.csv")

df.head()
```

```python
# Calculate Customer Lifetime Value (LTV)

df["LTV"] = df["MonthlyCharges"] * df["tenure"]

df[["MonthlyCharges", "tenure", "LTV"]].head()
```

```python
# Display important details of top 10 high-value customers

high_value_customers = df.sort_values(by="LTV", ascending=False)

high_value_customers[
    ["tenure", "MonthlyCharges", "TotalCharges", "LTV", "Churn_Yes"]
].head(10)
```

```python
# High-value customers who are at risk of churn

high_value_at_risk = high_value_customers[
    high_value_customers["Churn_Yes"] == 1
]

high_value_at_risk[
    ["tenure", "MonthlyCharges", "TotalCharges", "LTV", "Churn_Yes"]
].head(10)
```
