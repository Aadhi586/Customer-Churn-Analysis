```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("../Data/Telco_Customer_Churn_Dataset.csv")

df.head()
```

```python
print(df.columns)
```

```python
# Calculate and visualize overall churn rate

print("Customer Churn Count:")
print(df["Churn"].value_counts())

print("\nCustomer Churn Percentage:")
print((df["Churn"].value_counts(normalize=True) * 100).round(2))

plt.figure(figsize=(6,4))

sns.countplot(x="Churn", data=df)

plt.title("Overall Customer Churn Rate")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

```python
# Customer Distribution by Demographics

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.countplot(x="gender", data=df, ax=axes[0,0])
axes[0,0].set_title("Gender Distribution")

sns.countplot(x="SeniorCitizen", data=df, ax=axes[0,1])
axes[0,1].set_title("Senior Citizen Distribution")

sns.countplot(x="Partner", data=df, ax=axes[1,0])
axes[1,0].set_title("Partner Distribution")

sns.countplot(x="Dependents", data=df, ax=axes[1,1])
axes[1,1].set_title("Dependents Distribution")

plt.tight_layout()
plt.show()
```

```python
# Analyze Tenure Distribution

plt.figure(figsize=(8,5))

sns.histplot(df["tenure"], bins=30, kde=True, color="steelblue")

plt.title("Tenure Distribution of Customers")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")

plt.show()
```

```python
# Relationship between Churn and Contract Type / Payment Method

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Churn vs Contract Type
sns.countplot(x="Contract", hue="Churn", data=df, ax=axes[0])
axes[0].set_title("Churn by Contract Type")
axes[0].tick_params(axis='x', rotation=15)

# Churn vs Payment Method
sns.countplot(x="PaymentMethod", hue="Churn", data=df, ax=axes[1])
axes[1].set_title("Churn by Payment Method")
axes[1].tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```
